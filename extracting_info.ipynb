```python
from IPython import get_ipython
from IPython.display import display

from google.colab import files
upload_files = files.upload()


!pip install -q -U google-generativeai # Use the official google-generativeai library


import google.generativeai as genai # Import the library

# You need to configure your API key here. Replace 'YOUR_API_KEY' with your actual key.
# It's recommended to use environment variables for storing API keys.
# import os
# genai.configure(api_key=os.environ['GOOGLE_API_KEY'])
# Or directly configure it for this example:
genai.configure(api_key='YOUR_API_KEY') # Replace 'YOUR_API_KEY' with your actual key

# Create an instance of the GenerativeModel class
model = genai.GenerativeModel(model_name="gemini-2.0-flash") # Specify the model name when creating the instance

# Extract the filename from the uploaded files dictionary
if upload_files:
    filename = list(upload_files.keys())[0] # Assuming you only uploaded one file

# Use upload_file with the filename
# The upload_file method is a top-level method of the genai module, so calling genai.upload_file is correct here.
uploaded_file = genai.upload_file(path=filename) # Renamed variable to avoid confusion

prompt_text = "tell me what do you found in this image"
# prompt_text = "from which part of the world this species belongs"

# Call generate_content on the model instance, passing the uploaded file and prompt
response = model.generate_content(
    contents= [uploaded_file, prompt_text]
)
print("gemini says:", response.text)
```
